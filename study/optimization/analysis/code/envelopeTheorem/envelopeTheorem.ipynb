```python
# !pip install japanize-matplotlib
import os
from matplotlib import pyplot as plt
import japanize_matplotlib
```

```python
# 画像を生成

os.makedirs("./jpg", exist_ok=True)

envel_x = []
envel_y = []

plt.rcParams["mathtext.fontset"] = 'cm'

for param in range(-20, 21, 1):

    fig = plt.figure(figsize=(4,4))
    fig.subplots_adjust(left=0.18, bottom=0.18)

    ax1 = fig.add_subplot(1, 1, 1)

    ax1.set_xlabel(r"$x$", fontsize=16)
    ax1.set_ylabel(r"$y$", fontsize=16)

    ax1.set_xlim(-20, 20)
    ax1.set_ylim(-50, 450)

    ax1.minorticks_on()

    ax1.grid(which="major", color="#cccccc", linestyle="-")
    ax1.grid(which="minor", color="#eeeeee", linestyle="--")

    for param_2 in [-10] + list(range(-20, param, 1)):

        linear_x = [-20, 20]
        linear_y = [-40*param_2-param_2**2, 40*param_2-param_2**2]
        
        ax1.plot(linear_x, linear_y, color="#bbbbbb", linewidth=1)

    setten_x = [param]
    setten_y = [param**2]

    envel_x.append(setten_x[0])
    envel_y.append(setten_y[0])

    


    ax1.plot(linear_x, linear_y, color="#202020", linewidth=1)
    ax1.plot(setten_x, setten_y, color="#e25c00", marker="o", markersize=4)
    ax1.plot(envel_x, envel_y, color="#e25c00", linewidth=2)

    fig.savefig("./jpg/" + str(param+20) + ".jpg")

    if param == 20:
        plt.show()
    plt.close()
```
